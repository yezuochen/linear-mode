```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
football = pd.read_csv("b.1 football.txt", sep = " ")
print(football.head())
```

```python
x = football["x8"].values.reshape((-1,1))
y = football["y"].values.reshape((-1,1))
print(x)
print(y)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred, color = "black")

plt.title("Football Performance Scatter Plot and Regression")
plt.xlabel("Opponents")
plt.ylabel("Team")

plt.show()
```

```python
y_res = y - y_pred

sm.qqplot(y_res, fit = True, line = "45")
plt.title("Normal Probability Plot of the Residuals")
plt.show()
```

```python
plt.scatter(y_pred, y_res)
plt.axhline(y = 0, linestyle = "--", color = "black")

plt.title("Residual versus Fitted Value")
plt.xlabel("Fitted Value")
plt.ylabel("Deleted Residual")

plt.show()
```
